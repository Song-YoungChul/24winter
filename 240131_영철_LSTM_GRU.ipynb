```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


file = "emotion_corpus.xls"
f = pd.read_excel(file)
f
```

```python
df = f[["사람문장1","감정_대분류"]]
df
```

```python
# !pip install konlpy

from konlpy.tag import Okt
okt = Okt()
df['사람문장1'] = df['사람문장1'].map(lambda x: ' '.join(okt.morphs(x, stem = True)))
```

```python
def made_text(sent):
    x = okt.morphs(sent, stem = True)
    made_text = []
    for j in x:
        made_text.append(dict_lst[j])
    if len(made_text) < 67:
        for k in range(67-len(made_text)):
            made_text.append(99999)
    return made_text
```

```python
text = [i for i in df.iloc[:,0]]
text
```

```python
total_texts = ""
for i in text:
    total_texts = total_texts + " " + i
word_bag = pd.DataFrame(total_texts.split(" ")).iloc[:,0].unique().tolist()
word_bag
```

```python
dict_lst = dict(zip(word_bag, list(range(10812))))
dict_lst
```

```python
len(word_bag)
```

```python
# max([len(i.split(" ")) for i in text])
```

```python
emb = []
for i in text:
    tmp_text = []
    for j in i.split(" "):
        tmp_text.append(dict_lst[j])
    if len(tmp_text) < 67:
        for k in range(67-len(tmp_text)):
            tmp_text.append(99999)
    emb.append(tmp_text)
emb
```

```python
df["사람문장1"] = emb
```

```python
df.loc[(df["감정_대분류"] == "분노"),"감정_대분류"] = 0
df.loc[(df["감정_대분류"] == "기쁨"),"감정_대분류"] = 1
df.loc[(df["감정_대분류"] == "불안"),"감정_대분류"] = 2
df.loc[(df["감정_대분류"] == "당황"),"감정_대분류"] = 3
df.loc[(df["감정_대분류"] == "슬픔"),"감정_대분류"] = 4
df.loc[(df["감정_대분류"] == "상처"),"감정_대분류"] = 5
df.columns=["document","label"]
df
# df["감정_대분류"].unique()
```

```python
emotion = [i for i in df.iloc[:,1]]
total_emotion = []
for i in emotion:
    c = []
    for j in range(6):
        if j == i:
            c.append(1)
        else:
            c.append(0)
    total_emotion.append(c)
total_emotion
```

```python
df["label"] = total_emotion
df
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

train_texts = train_df['document'].tolist() # 문자열 데이터로 명시 후 리스트 화
train_labels = train_df['label'].tolist()
test_texts = test_df['document'].tolist()
test_labels = test_df['label'].tolist()
print(len(train_texts), len(test_labels))
```

```python
from numpy import array

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, GRU, Dropout, Activation
import tensorflow as tf

import numpy as np
```

```python
model = Sequential()

model.add(LSTM(64, return_sequences = True, input_shape=(67, 1)))

model.add(LSTM(128, return_sequences = True))

model.add(LSTM(64, return_sequences = False))

model.add(Dense(6, activation = "softmax"))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)

model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics = ['accuracy'])

model.summary()

hist = model.fit(train_texts, train_labels, validation_data = (test_texts, test_labels), batch_size=5, epochs=5)
```
